```python
import pandas as pd
import numpy as np
```

```python
#Cadastro da Loja A
cadastro_a = {'id': ['AA2', 'BB4', 'CC3', 'DD4', 'EE5', 'FF6'],
              'Nome': ['Arlindo', 'Betuel', 'Carlos', 'Danilo', 'Ester', 'Fabio'],
              'Idade': [20,15,54,65,35,44],
              'CEP': ['111-111','222-222','333-333','444-444','555-555','666-666']  
              }

cadastro_a = pd.DataFrame(cadastro_a, columns=['id', 'Nome', 'Idade', 'CEP'])
cadastro_a
```

```python
#Cadastro da Loja B
cadastro_b = {'id': ['GG7', 'HH8', 'CC3', 'JJ9', 'II10', 'FF6'],
              'Nome': ['Gabriel', 'Helio', 'Carlos', 'Jamelo', 'Ingrid', 'Fabio'],
              'Idade': [24,25,54,47,28,44],
              'CEP': ['777-111','888-222','333-333','999-444','1010-555','666-666']  
              }
cadastro_b = pd.DataFrame(cadastro_b, columns=['id', 'Nome', 'Idade', 'CEP'])
cadastro_b
```

```python
#Registro de compras de todas unidades
compras = {'id': ['AA2', 'EE5', 'CC3', 'EE5', 'CC3', 'FF6', 'HH8','HH8'],
              'Data': ['2022-10-12', '2022-11-02', '2022-12-01', '2022-12-25', '2023-07-13', '2022-11-07', '2022-02-22', '2023-01-01'],
              'Valor': [200,100,40,150,300,25,50,500],
              }
compras = pd.DataFrame(compras, columns=['id', 'Data', 'Valor'])
compras
```

```python
inner = pd.merge(cadastro_a,cadastro_b[['id','Idade','CEP']], on=['id'], how='inner', suffixes=('_A','_B'))
inner
```

```python
full_join_concat = pd.concat([cadastro_a,cadastro_b], ignore_index=True)
full_join_concat.drop_duplicates(subset='id')
```

```python
compras_a = pd.merge(cadastro_a,compras, on=['id'], how='left')
compras_a
```

```python
compras_b = pd.merge(cadastro_b,compras, on=['id'], how='left')
compras_b
```

```python
compras_b.groupby(['id', 'Nome'])['Valor'].sum() #mean()
```

```python
# Outer
outer = pd.merge(cadastro_b,cadastro_a, on=['id'], how='outer', indicator=True)
outer
```

#### Group By

```python

df = pd.DataFrame({
    'A': ['verdadeiro', 'falso', 'verdadeiro', 'falso', 'verdadeiro', 'falso','verdadeiro', 'falso'],
    'B': ['um', 'um', 'dois', 'tres', 'dois', 'dois', 'um', 'tres'],
    'C': abs(np.random.randn(8) *10),
    'D': abs(np.random.randn(8) *10)
})

df
```

```python
df.groupby(['A']).sum() #mean()
```

```python
df.groupby(['B']).sum() #mean()
```

```python
df.groupby(['A','B']).mean()
```
